## Gregory-Leibniz Series
This is used to calculate $\pi$

```julia
using Plots
```

According to the series
$$\frac{\pi}{4}=1-\frac{1}{3}+\frac{1}{5}-\frac{1}{7}+\frac{1}{9}-\ldots$$

```julia
function calculator(n)
    pi_value,flag_coeff = 1,-1
    i=1
    while i<n
        pi_value = pi_value + (flag_coeff^i)*1/(2*i+1)
        i=i+1
        flag_coeff=flag_coeff*-1
    end
    return 4*pi_value
end
```

##### Printing the values of $\pi$ below

```julia
println("Value using 5 terms = ",calculator(5))
println("Value using 10 terms = ",calculator(10))
println("Value using 15 terms = ",calculator(15))
```

Effect of adding more number of terms

```julia
dif=[]
pi_values = []
pi_even,pi_odd=[],[]
even_x,odd_x=[],[]
anim = @animate for n in 1:200
    push!(pi_values,calculator(n))
    if length(pi_values)%2==0
        push!(pi_even,pi_values[end])
        push!(even_x,n)
    else
        push!(pi_odd,pi_values[end])
        push!(odd_x,n)
    end
    if n>1
        push!(dif,pi_values[n]-pi_values[n-1])
    else
        push!(dif,0)
    end
    plot(pi_values,xlim=(1,200),ylim=(3,4),label=false)
    scatter!(even_x,pi_even,xlim=(1,200),ylim=(3,4),label=false)
    scatter!(odd_x,pi_odd,xlim=(1,200),ylim=(3,4),label=false)
    title!(string(abs(dif[end])))

end
gif(anim,"PiCalculator.gif",fps=30)
```
