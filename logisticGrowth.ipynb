```python
from scipy.integrate import odeint
import numpy
import matplotlib.pyplot as plt
```

## Logistic Growth
<b>Introduction:</b> Exponential growth may happen for a while, if there are few individuals and many resources. But when the number of individuals gets large enough, resources start to get used up, slowing the growth rate. Eventually, the growth rate will plateau, or level off, making an S-shaped curve. The population size at which it levels off, which represents the maximum population size a particular environment can support, is called the carrying capacity, or K.<br>
The differential Equation that governs the model is given by $\frac{\mathrm{d} N}{\mathrm{d} t} = r\times N\times \frac{N-K}{K}$<br>
wherein <br><t>r -> Per Capita Growth Rate<br>
    <t>K -> Carrying Capacity<br>
        <t>N -> Population<br>
           

```python
def logGrowthModel(N,t):
    r = 0.08 #Per capita Growth Rate (8 percent)
    K = 1000 #how close it is to carrying capacity
    dNdt = r*N*(1-N/K) #Derivative that governs the growth or by definition
    return dNdt
```

So (K-N) will give us how many more individuals can be added to the population before it hits carrying capacity

```python
t = numpy.linspace(0,150,1000)
N0=10 #If initial Number is 10
N=odeint(logGrowthModel,N0,t) #Function to fit the values, solve the differential equation and return N values
```

```python
plt.title('Logistic Growth')
plt.xlabel('Time ->')
plt.ylabel('Population ->')
plt.plot(t,N)
plt.show()
```

Going deeper into the topic<br>
We shall change the initial number and see the difference in the curve we get

```python
def CalculateAndPlot(N0,t):
    N=odeint(logGrowthModel,N0,t)
    plt.plot(t,N,label='Ini. population: '+str(N0))

t = numpy.linspace(0,150,1000)
plt.title('Logistic Growth')
plt.xlabel('Time ->')
plt.ylabel('Population ->')
CalculateAndPlot(0,t)
CalculateAndPlot(10,t)
CalculateAndPlot(50,t)
CalculateAndPlot(100,t)
plt.legend()
plt.show()
```

Inference: More is the initial population, faster it does reaches the saturation for other parameters fixed

##### Inflection Point:
Inflection point is a point on the curve where the slope is steepest. In other words, It gives the point after which the growth rate starts reducing below 1.<br>
It is found out by equating second deravative of N to 0 and solving for it<br>
It is given by : <br>
$(\log (\frac{K - N0}{N0}) , \frac{K}{2})$


```python
import math
t = numpy.linspace(0,150,1000)
plt.title('Logistic Growth')
plt.xlabel('Time ->')
plt.ylabel('Population ->')
CalculateAndPlot(10,t)
plt.plot(t,numpy.array(1000/2 * numpy.ones(len(t))),linestyle='dashed', label='Inflection point')
plt.legend()
plt.show()
```
